```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("images/lena_color.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

#img_flip_lr = cv2.flip(img, 1) #left-right flip
img_flip_ud = cv2.flip(img, 0) #up-down flip
plt.imshow(img_flip_ud)
plt.title('left-right flip')
```

```python
#rotation
output_ROTATE = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
#output_ROTATE = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
plt.figure()
plt.imshow(output_ROTATE)
plt.title('ROTATE_90_CLOCKWISE')
```
